```python
class sample_class:
  var0 = "クラス変数の文字列" # イミュータブルなクラス変数
  var1 = ["クラス変数のリスト"] # ミュータブルなクラス変数
  
  def __init__(self, var2):
    self.var2=var2 # メンバ変数
  def add_var1(self, value):
    self.var1 += [value]

print("\n[DEBUG] a")
print(sample_class.var0) # アクセスできる
print(sample_class.var1) # アクセスできる
# print(sample_class.var2) # インスタンス化しないとアクセスできない

# インスタンス化
instance = sample_class(var2="メンバ変数")
instance_2 = sample_class(var2="メンバ変数です！")

print("\n[DEBUG] b-1")
print(instance.var0) # アクセスできる
print(instance.var1) # アクセスできる
print(instance.var2) # アクセスできる
print("\n[DEBUG] b-2")
print(instance_2.var0) # アクセスできる
print(instance_2.var1) # アクセスできる
print(instance_2.var2) # アクセスできる

# 値の変更
instance.var0="クラス変数の文字列を変更してみました"
instance.var1=["クラス変数の文字列を変更してみました"]
instance.var2="メンバ変数の文字列を変更してみました"
instance.add_var1(value="クラス変数のリストに追加です")

#instance_2.var0='変更不可？'

print("\n[DEBUG] c-1")
print(instance.var0) # 変更される
print(instance.var1) # 変更される
print(instance.var2) # 変更される
print("\n[DEBUG] c-2")
print(instance_2.var0) # イミュータブルなクラス変数は変更されない!!!
print(instance_2.var1) # ミュータブルなクラス変数は変更される!!!
print(instance_2.var2) # 変更されない
#折りたたむ

sample_class.var0='変更'
print(sample_class.var0)

```
